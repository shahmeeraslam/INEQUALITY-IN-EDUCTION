 Trend Analysis: Change in Inequality Over Time

To understand how educational inequality has evolved, we compared the average inequality values for the years 2010 and 2021.

Average Inequality in 2010:
25.93

Average Inequality in 2021:
23.73

 Conclusion

 The results clearly indicate that educational inequality has decreased over time.
Despite several countries still facing high inequality levels, the overall global trend shows improvement between 2010 and 2021.

```python
data["trend direction"] = filt["Inequality in Education (2010)"] - filt["Inequality in Education (2021)"]
def trend_lebal(x):
    if x > 0.5:
        return "Increasing"
    elif x < -0.5:
        return "Decreasing"
    else:
        return "Stable" 
data["trend_direction"] = data["trend direction"].apply(trend_lebal)
sort_data = data[["Country","avg_ineq_c","trend_direction"]].sort_values(by="avg_ineq_c",ascending=False)
print(sort_data.to_string())
```
